```python
import pandas as pd

weather_api_data = pd.read_csv("weather_data.csv")
weather_api_data.head()
```

```python
print(weather_api_data.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Relation entre la temprature en Celsius et en Fahrenheit
sns.scatterplot(x='TemperatureC', y='TemperatureF', data=weather_api_data)
plt.title('Relation entre la temprature en Celsius et Fahrenheit')
plt.show()

# Distribution de la temprature en Celsius
sns.histplot(data=weather_api_data, x='TemperatureC', bins=20, kde=True)
plt.title('Distribution de la temprature en Celsius')
plt.show()

# Condition mtorologique par emplacement
plt.figure(figsize=(10, 6))
sns.countplot(data=weather_api_data, x='Condition', hue='Location')
plt.xticks(rotation=45)
plt.title('Rpartition des conditions mtorologiques par emplacement')
plt.legend(loc='upper right')
plt.show()

```

```python
# Convertir les colonnes de temps en objets datetime
weather_api_data['LocalTime'] = pd.to_datetime(weather_api_data['LocalTime'])
weather_api_data['LastUpdated'] = pd.to_datetime(weather_api_data['LastUpdated'])

# Visualisation de la temprature au fil du temps
plt.figure(figsize=(10, 6))
sns.lineplot(x='LocalTime', y='TemperatureC', data=weather_api_data, hue='Location')
plt.title('Tendance de la temprature au fil du temps')
plt.xticks(rotation=45)
plt.show()
```

```python
# Histogramme de la temprature en Celsius
plt.figure(figsize=(8, 6))
sns.histplot(data=weather_api_data, x='TemperatureC', bins=15, kde=True)
plt.title('Distribution de la temprature en Celsius')
plt.xlabel('Temprature en Celsius')
plt.ylabel('Frquence')
plt.show()
```
